# PySpark Hello World

The simplest possible PySpark program to get you started.

```python
from pyspark.sql import SparkSession

# Create a Spark session
spark = SparkSession.builder \
    .appName("Hello World") \
    .master("local[*]") \
    .getOrCreate()

print("Hello, PySpark World!")
print(f"Spark version: {spark.version}")
print(f"Running on: {spark.sparkContext.master}")
```

```python
# Create a simple DataFrame
data = [("Hello",), ("PySpark",), ("World",)]
df = spark.createDataFrame(data, ["message"])

print("\nOur first DataFrame:")
df.show()
```

```python
# Perform a simple transformation
from pyspark.sql import functions as F

df_upper = df.withColumn("message_upper", F.upper(F.col("message")))
df_upper.show()
```

```python
# Count records
count = df.count()
print(f"\nTotal records: {count}")
```

```python
# Stop the Spark session
spark.stop()
print("\nGoodbye, PySpark World!")
```
